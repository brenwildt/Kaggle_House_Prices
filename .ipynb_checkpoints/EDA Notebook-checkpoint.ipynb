```python
num_df = df.select_dtypes(include=['number'])
```

```python
num_df['TotalSQFT'] = num_df['1stFlrSF'] + num_df['2ndFlrSF'] + num_df['TotalBsmtSF']

#Correlation Matrix 
corr_matrix = num_df.corr()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Numeric Features')
plt.show
```

```python
threshold = 0.6  # Example threshold for high correlation

# Identifying high correlations without duplicates and self-correlations
high_corr_pairs = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
filtered_pairs = high_corr_pairs.stack().reset_index()
filtered_pairs.columns = ['Feature1', 'Feature2', 'Correlation']
filtered_pairs = filtered_pairs[abs(filtered_pairs['Correlation']) > threshold]
```

```python
# Create a pivot table
pivot_table = filtered_pairs.pivot(index='Feature1', columns='Feature2', values='Correlation')
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of High Correlation Pairs')
plt.show()
```

```python
num_df = num_df.dropna(axis=1)
```

```python
#Initialize the scaler
scaler = StandardScaler()

#Scale the training data
num_df_scaled = scaler.fit_transform(num_df)
```

```python
#Separate the target variable from the features and perform the train/test split of the training data
X = num_df.drop('SalePrice', axis=1)
y = df['SalePrice']

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=42)
```

```python
#Initialize the Lasso regression model
lasso = Lasso(alpha=1.0)

#Fit the model
lasso.fit(X_train, y_train)
```

```python
#Evaluate the model by making predictions and generating an RMSE value
y_pred = lasso.predict(X_test)

rmse = root_mean_squared_error(y_test, y_pred)
print(f'Root Mean Squared Error: {rmse}')
```

```python
#Get feature coefficients that have non-zero values and print the names of those features
feature_names = X_train.columns

coefficients = lasso.coef_

selected_features = feature_names[coefficients != 0]

print('Selected Features and their Coefficients:')
for feature, coef in zip(feature_names, coefficients):
    if coef != 0:
        print(f'{feature}: {coef}')
```
